```python
from rdflib import Graph, URIRef

# Load your graph
g = Graph()
g.parse("../generated-rdf/timeseriesmapping.ttl", format="turtle")

# Prefix map
prefixes = {
    "ex": "http://example.com/",
    "rdfs": "http://www.w3.org/2000/01/rdf-schema/",
    "sosa": "http://www.w3.org/ns/sosa/",
    "ssn": "http://www.w3.org/ns/ssn/",
    "geo": "http://www.opengis.net/ont/geosparql/",
    "dcterms": "http://purl.org/dc/terms/",
    "qudt": "http://qudt.org/schema/qudt/",
    "unit": "http://qudt.org/vocab/unit/",
    "gn": "http://www.geonames.org/ontology/",
    "xsd": "http://www.w3.org/2001/XMLSchema/",
    "cdt": "https://ci.mines-stetienne.fr/lindt/v4/custom_datatypes#ucum/"
}

# Bind the prefixes to the graph's namespace manager (needed for qname)
for prefix, uri in prefixes.items():
    g.namespace_manager.bind(prefix, uri)

# Create a new graph to hold updated triples
new_graph = Graph()
for prefix, uri in prefixes.items():
    new_graph.namespace_manager.bind(prefix, uri)

# Replace subjects
for subj, pred, obj in g:
    new_subj = subj
    if isinstance(subj, URIRef):
        try:
            qname = g.namespace_manager.qname(subj)  # e.g., "ssn:FeatureOfInterest"
            new_subj = URIRef(f"http://example.com/{qname}")  # create a new URI
        except:
            pass  # if qname can't be resolved, keep the original

    new_graph.add((new_subj, pred, obj))

# Optional: save the updated graph
new_graph.serialize(destination="updated_subjects.ttl", format="turtle")

# Print some updated triples
for s, p, o in list(new_graph)[:5]:
    print(f"{s} -- {p} --> {o}")
```
